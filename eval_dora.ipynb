```python
from audiocraft.models import MusicGen
musicgen = MusicGen.get_pretrained(f'/home/thomasyim/cs229-music/content/checkpoints/finetune/{sid}')
# musicgen = MusicGen.get_pretrained('facebook/musicgen-large')
```

```python
from audiocraft.data.audio import audio_write
musicgen.set_generation_params(duration=10)  # generate 8 seconds.
# wav = musicgen.generate_unconditional(4)    # generates 4 unconditional audio samples
descriptions = ['a techno beat']
wav = musicgen.generate(descriptions)  # generates 3 samples.

for idx, one_wav in enumerate(wav):
    # Will save under {idx}.wav, with loudness normalization at -14 db LUFS.
    audio_write(f'{idx}', one_wav.cpu(), musicgen.sample_rate, strategy="loudness", loudness_compressor=True)
```
